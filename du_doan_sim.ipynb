```python
#import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
traindataset_df=pd.read_csv('train_dataset.csv')
```

```python
traindataset_df.head()
```

```python
traindataset_df.describe()
```

```python
fields = ['sim_number','price_vnd','price_dollar']
traindataset_df_new=traindataset_df.drop(fields, axis = 1)
traindataset_df_new.head()
```

```python
#corr()
traindataset_corr = traindataset_df_new.corr()
traindataset_corr
```

```python
#one_like
ones_corr = np.ones_like(traindataset_corr
, dtype = bool)
ones_corr
```

```python
ones_corr.shape,traindataset_corr.shape
```

```python
# np's triu
mask = np.triu(ones_corr)
mask
```

```python
fig, ax = plt.subplots(figsize=(12,10))
sns.heatmap(data =traindataset_corr, mask = mask, annot = True, cmap ='Blues', annot_kws ={"fontsize":6}, fmt=".4f", linewidths = 0.5)
```

```python
for col in traindataset_df.columns:
    missing_data = traindataset_df[col].isna().sum()
    missing_percent = missing_data/len(traindataset_df)*100
    print(f"Column {col}: has {missing_percent}% missing data")
```

```python
fig, ax =plt.subplots(figsize=(8,5))
sns.heatmap(traindataset_df.isna(), cmap="Blues", cbar = False, yticklabels = False)
```

```python
features = ["sim_luc_quy","sim_ngu_quy","sim_tu_quy","sim_tu_quy_giua","sim_luc_quy_giua","sim_tam_hoa","sim_tam_hoa_giua","sim_tam_hoa_kep","sim_nam_sinh","sim_kep","sim_lap","sim_taxi2","sim_taxi3","sim_taxi4","sim_so_la_loai1","sim_so_la_loai2","sim_so_la_loai3"]
```

```python
x = traindataset_df[features]
y = traindataset_df["price_dollar"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x,y,train_size = 0.9,test_size =0.1, random_state =0)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
dt_model = LinearRegression()
```

```python
dt_model.fit(x_train,y_train)
```

```python
y_preds = dt_model.predict(x_valid)
```

```python
y_preds
```

```python
pd.DataFrame({'y':y_valid,'y_preds':y_preds})
```

```python
plt.scatter(y_valid,y_preds)
```

```python
sns.distplot(y_valid - y_preds)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_valid, y_preds))
print('MSE:', metrics.mean_squared_error(y_valid, y_preds))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_valid, y_preds)))
```

```python
metrics.explained_variance_score(y_valid, y_preds)
```

```python
coeff_df = pd.DataFrame(dt_model.coef_,x.columns,columns=['Coefficient'])
coeff_df
```
